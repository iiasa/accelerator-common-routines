## 1. Fetch GitHub App Token

```python
import requests
TOKEN_API = "https://your-backend.com/get_github_app_token"
response = requests.get(TOKEN_API, params={"user_id": "123"})
token = response.json().get("token")
token
```

## 2. Store Git Credentials

```python
import os
os.system("git config --global credential.helper store")
with open(os.path.expanduser("~/.git-credentials"), "w") as f:
    f.write(f"https://x-access-token:{token}@github.com\n")
print("Token stored.")
```
